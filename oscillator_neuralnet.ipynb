Se resuelve la ecuación diferencial

$\frac{{d^2y}}{{dt^2}} + \omega^2 y = 0$


```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Se define la función de la ecuación diferencial del oscilador armónico simple
def oscillator(y, t, omega):
    dydt = [y[1], -omega**2 * y[0]]
    return dydt

# Función para resolver la ecuación diferencial con parámetro omega variable
def solve_differential_equation(omega_values, y0, t):
    results = []
    for omega in omega_values:
        sol = odeint(oscillator, y0, t, args=(omega,))
        results.append(sol)
    return results

# Condiciones iniciales
y0 = [1.0, 0.0]  # posición inicial y velocidad inicial

# Definir el intervalo de tiempo en el que se resolverá la ecuación
t = np.linspace(0,10,100)

# Valores para el parámetro omega
omega_values = [1.0,2.0, 3.0]
# omega_values = [1.0, 2.0]
# omega_values = [1.0]

# Resolver la ecuación diferencial con diferentes valores de omega
results = solve_differential_equation(omega_values, y0, t)
```

```python
results=[]
# for i in t:
for omega in omega_values:
  sol = solve_differential_equation(np.array([omega]), y0, t)
  results.append(sol)
  # print(omega, sol)
```

```python
dataset = []
for i, sol_om in enumerate(results):
  # print(sol_om)
  # print(np.shape(sol_om))
  sol_om = np.squeeze(sol_om)
  # print(np.shape(sol_om))
  for j, sol_ti in enumerate(sol_om):
    # print("Omega: {} t: {} sol[0]: {} sol[1]: {}".format(omega_values[i], t[j], sol_ti[0], sol_ti[1]))
    dataset.append([omega_values[i], t[j], sol_ti[0], sol_ti[1]])
```
