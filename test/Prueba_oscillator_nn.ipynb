```python
import matplotlib.pyplot as plt
from tensorflow import keras
import numpy as np

# Carga el modelo desde el archivo .h5
model = keras.models.load_model("/content/drive/MyDrive/Colab Notebooks/saved_models/oscillator_nn.h5")

```

```python
import random

# Definir el tamaño del arreglo
filas = 500  #Número de filas

# Crear una lista de listas para almacenar los valores
x = []
x2 = []
# Generar el arreglo con números aleatorios en formato de punto flotante
for _ in range(filas):
    columna1 = random.choice([2.0])
    columna2 = random.uniform(0, 10)
    x.append([columna1, columna2])

for _ in range(filas):
    columna1 = random.choice([1.0])
    columna2 = random.uniform(0, 10)
    x2.append([columna1, columna2])

# Imprimir el arreglo generadorim
# for fila in New_X:
#     print(fila)
```

```python
x = np.array(x)
np.shape(x)
time = x[:,1]


x2 = np.array(x2)
np.shape(x)
time2 = x2[:,1]
# pred = predictions[1]

# print(predictions)
# plt.scatter(time,pred, color = 'blue')
```

```python
# Prediccion utilizando el modelo
predictions = model.predict(x)
predictions2 = model.predict(x2)
# print(predictions)
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(time,predictions[:,1], color = 'blue')
plt.scatter(time2,predictions2[:,1], color = 'red')
plt.xlabel('Tiempo')
plt.ylabel('Posición')

plt.legend()
plt.grid(True)
plt.show()

```
